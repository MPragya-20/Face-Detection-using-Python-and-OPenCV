# Import necessary Libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

# Read the image file

```python
img = cv2.imread('School.jpg')
```

# Since cv2 interprets BGR format, make sure to convert it to RGB format

```python
plt.imshow(img) #BGR color format displayed
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

# Get haar_cascade_frontal_face classifier(XML file)

```python
img_h_cascade= cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

# Use detectMultiScale() function under the classifier to detect face with rectangles around

```python
rect_detect = img_h_cascade.detectMultiScale(image = rgb_img,scaleFactor = 1.05,minNeighbors = 8)
```

```python
for (x_c,y_c,width,height) in rect_detect:
    cv2.rectangle(rgb_img , (x_c,y_c),(x_c+width , y_c+height),(255,255,0),4)
```

```python
plt.imshow(rgb_img)
```
